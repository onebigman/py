```python
myset = {1, 2} # Directly assigning values to a set
print(myset)
myset = set()  # Initializing a set
print(myset)
myset = set(['a', 'b']) # Creating a set from a list
print(myset)
myset.add(('a',5))
print(myset)
myset.update(('a',7))
print(myset)
```

```python
>> a = {2, 4, 5, 9}
>> b = {2, 4, 11, 12}
>> a.union(b) # Values which exist in a or b
{2, 4, 5, 9, 11, 12}
>> a.intersection(b) # Values which exist in a and b
{2, 4}
>> a.difference(b) # Values which exist in a but not in b
{9, 5}
```

```python
# *的使用
from itertools import permutations
a = [1, 2, 3]
c = ['h','a','c','k']
b = list(permutations(c,2))

print(*b,sep='\n')
print(*c)
[print(p,sep='')for p in list(permutations(c,2))]
[print(*p,sep='')for p in list(permutations(c,2))]
print('------')
print(*[''.join(i) for i in permutations(c,2)],sep='\n')
print('------')
print(*[''.join(i) for i in list(permutations(c,2))],sep='\n')
print('------')
print([''.join(i) for i in permutations(sorted(c),2)],sep='\n')
print(i for i in list(permutations(sorted(c),2)))
print(*a)
print(*a,sep='')
for p in permutations(c,2):
    print(''.join(p))
```
